1. Update `Tavg` Missing Data by taking average of `Tmax` and `Tmin` and change `Tavg` datatype to float.

```python
weather[weather['Tavg']=='M']
# index row 7, 505, 675, 1637, 2067, 2211, 2501, 2511, 2525, 2579,2811
```

```python
weather.loc[weather['Tavg'] == 'M', 'Tavg'] =round(((weather['Tmax']+weather['Tmin'])/2))
```

```python
# check result after update. Tavg was not 'Round-Up', just roudn to nearest int
weather.iloc[[ 7, 505, 675, 1637, 2067, 2211, 2501, 2511, 2525, 2579,2811],0:5]
```

```python
weather['Tavg']=pd.to_numeric(weather['Tavg'])
print(weather['Tavg'].dtypes)
```

2. Create a df without date duplication.

```python
station_1=weather[weather['Station']==1]
station_2=weather[weather['Station']==2]
combined=station_1.merge(station_2,on='Date',how='inner',suffixes=('_1','_2'))
print(combined.shape)
combined.head(2)
```

3. Update `Heat`,`Cool`Missing Data in Stn 2 with information from Stn 1 & change data type to int

```python
check_heat_cool=combined[(combined['Heat_2']=='M')|(combined['Cool_2']=='M')]
check_heat_cool_miss=check_heat_cool[['Station_1','Station_2','Heat_1','Heat_2','Cool_1','Cool_2']]
check_heat_cool_miss
#iloc=[3,252,337,818,1033,1105,1250,1255,1262,1289,1405]
```

```python
combined.loc[combined['Heat_2']=='M','Heat_2']=combined['Heat_1']
combined.loc[combined['Cool_2']=='M','Cool_2']=combined['Cool_1']
```

```python
# Check after amendment
combined.iloc[[ 3,252,337,818,1033,1105,1250,1255,1262,1289,1405],[0,8,9,22,29,30]]
```

```python
# change datatype for Heat_1','Heat_2','Cool_1','Cool_2'

combined[['Heat_1','Heat_2','Cool_1','Cool_2']]=combined[['Heat_1','Heat_2','Cool_1','Cool_2']].apply(pd.to_numeric)
print(combined['Heat_1'].dtypes)
print(combined['Heat_2'].dtypes)
print(combined['Cool_1'].dtypes)
print(combined['Cool_2'].dtypes)
```
